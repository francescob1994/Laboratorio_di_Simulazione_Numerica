## Simulated Annealing

In questa lezione ho modficato il codice della lezione precedente in modo da risolvere il TSP con un Simulated Annealing.

Si stabilisce un set di valori di beta $\{\beta _1, \cdots, \beta _N\}$ e un numero $N$ di generazioni per ciascun valore di $\beta$. Ad ogni generazione si costruisce la successiva mediante un algoritmo Metropolis che utilizza la temperatura fittizia $1/\beta$.

### Algoritmo Metropolis

L'algoritmo Metropolis ora opera nel modo seguente:
- siano $X$, $Y$ i genitori scelti con $l_{X} < l_{Y}$. Da questi si costruiscono i figli $XX$ e $YY$ con la funzione Crossover, come nel programma della lezione precedente;
- considero il figlio $XX$. Se $l_{XX} < l_{X}$ accetto $XX$ nella popolazione della generazione successiva. Altrimenti lo accetto con probabilità $p=\exp( -\beta\cdot l_{XX})/\exp( -\beta\cdot l_{X})$. In caso di rigetto viene mantenuto il cromosoma $X$ nella nuova popolazione;
- considero il filgio $YY$. Se $l_{YY} < l_{X}$ accetto $YY$ nella popolazione della generazione successiva.
Altrimenti lo accetto con probabilità $p=\exp( -\beta\cdot l_{YY})/\exp( -\beta\cdot l_{X})$. In caso di rigetto se nel punto precedente ho accettato $XX$ ora aggiungo $X$ alla nuova popolazione, altrimenti aggiungo $YY$.

## Parallel Simulated Annealing

In questo paragrafo vediamo un ulteriore miglioramento del codice che consiste nel dividere l'ottimizzazione su più core. 
Come generatore di numeri casuali questa volta, anzichè la funzione drand48(), ho utilizzato l'oggetto Random. Inoltre ciascun core legge una riga diffenete del file Primes in modo da iniziailizzare l'oggetto Random in modo diverso e produrre sequenze di numeri pseudocasuali differenti.

Ogni core esegue in parallelo la simualzione e alla fine di ogni iterazione (crezaione di una nuova popolazione) tutti i core inviano al core di rank zero la lunghezza $l_{\text{min}}$ del cromosoma di lunghezza minima. Il core zero utilizza la funzione SortPrint per selezionare il minimo di questi $l_{\text{min}}$ e lo stampa su file.
Nella figura seguente osserviamo l'andamento di questa lunghezza minima con il numero di generazioni (utilizzando città sulla circonferenza).


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

t, lmin = np.loadtxt("MPI/length.dat", usecols=(0,1), unpack = True )
def f(x,a,l,b):
   return a*np.exp(-x/l)+b
plt.figure(figsize=[16, 8])
p_opt, p_cov = curve_fit(f, t, lmin)
y_fit = f(t,p_opt[0],p_opt[1],p_opt[2])
plt.subplot(121)
plt.plot(t,y_fit,label='fitting function')
plt.plot(t,lmin, label='l_min')
plt.legend()
plt.title("tau = "+str(int(p_opt[1])))
plt.xlabel('generation')
plt.ylabel('l_min')
plt.grid(True)

plt.show()
```

Dopo l'ultima iterazione il core cui corrisponde il cromosoma di lunghezza minima (tra i cromosomi di tutti i core) stampa la sequanza di città associata. Di seguito vediamo il percorso migliore trovato nel caso di città sulla circonferenza: 

```python
data = np.loadtxt("FinalSeq.dat", usecols=(0))

b = data[0]
ll = data[1]
l= data[2]
seq = np.zeros(len(data)-3)
for i in range(len(data)-3):
   seq[i] = data[i+3]

C = np.loadtxt("circumference.dat", usecols=(0,1))
C1 = np.zeros((C.shape[0]+1, C.shape[1]))

for i in range(len(seq)):
   ind = int(seq[i])
   C1[i,:] = C[ind,:]

C1[len(seq),:]=C1[0,:]

plt.figure(figsize=[8, 8])
plt.scatter(C1[:,0],C1[:,1], label='cities')
plt.plot(C1[:,0],C1[:,1])
plt.legend()
plt.grid(True)
plt.title('Best path: L1 = '+str(l))
plt.show()
```

In queste simulazioni viene fissato un numero totale di cromosomi e ciascun core esegue una una simulazione aggiornando popolazioni contenenti un numero di cromosomi pari a (numero totale di cromosomi)/(numero di core).
Lasciando il numero totale di cromosomi fissato uguale a 600, ho osservato come varia il tempo di esecuzione in 
funzione del numero di core su cui viene lanciata la simulazione.
Per ciascun numero di core ho eseguito 50 simulazioni e calcolato la media $\langle t_{\text{exe}} \rangle$ e incertezza sulla media del tempo di esecuzione. 
Di seguito è illustrato l'andamendo di $\langle t_{\text{exe}} \rangle$ con il numero dei core. 

```python
cores, texe, err = np.loadtxt("MPI/mean.dat", usecols=(0,1,2), unpack=True)
plt.errorbar(cores, texe, yerr=err)
plt.grid(True)
plt.title('Execution Time')
plt.xlabel('#cores')
plt.ylabel('$<t_{exe}>$')
plt.scatter(cores,texe)
plt.show()
```

Come atteso il tempo di esecuzione diminuisce con il numero di core.
