# Lezione 08


 Si consideri una particella quantistica 1D confinata nel potenziale esterno $V(x) = x^4 - \frac{5}{2}x^2$.
L'obiettivo di questa lezione è ottimizare il ground state di tale sistema utilizzando un Variational Monte
Carlo. 
La funzione d'onda del ground state verrà approssimata con la funzione d'onda di prova $\psi ^{\mu,\sigma}(x) = e^{(x-\mu)^2/2\sigma ^2}+e^{(x+\mu)^2/2\sigma ^2}$ dove $\mu$ e $\sigma$ sono i parametri da ottimizzare. 

Per ottenere i valori di $\mu$ e $\sigma$ che minimizzano l'energia della particella considero 21 valori di $\mu$ nell'intervallo $[0.4,1.4]$ e 21 di $\sigma$ in $[0.1,1.1]$. Per ogni coppia di valori viene eseguita una simulazione Monte Carlo in cui viene stimato 
$$
E^{\mu,\sigma}_0=
\langle \psi ^{\mu,\sigma} |{\hat H}|\psi ^{\mu,\sigma} \rangle = 
\int dx \frac{|\psi ^{\mu,\sigma}(x)|^2}{\int dx |\psi ^{\mu,\sigma}(x)|^2} \frac{{\hat H} \psi ^{\mu,\sigma}(x)}{\psi ^{\mu,\sigma}(x)} = \int dx \rho(x) f(x).
$$

## Simulazione Monte Carlo

In ciascuna simulazione viene costruito un campione di $\rho(x)$ utilizzando l'algoritmo Metropolis:
scegliendo $x_0=0$ come punto iniziale, dato il punto $x_i$, si propone come punto successivo $x_{i+1}=x_i + dx$ con $dx\sim U(-\delta, \delta)$; se $\rho(x_{i+1})>\rho(x_{i})$ si accetta altrimenti si accetta con probabilità $\rho(x_{i+1})/\rho(x_i)$.

In un primo momento ho posto $\delta = 1$ e ho notato che ciò causava diverse percentuali di accettazione (tra il $40\%$ e $80\%$) a seconda dei valori di $\mu$ e $\sigma$. Successivamente ho provato a utilizzare $\delta = k + \sigma$ per diversi valori di $k$ nel tentativo di ottenere un accettazione meno variabile e intorno al $50\%$. Ciò non ha prodotto particolari miglioramenti così ho mantenuto $\delta = 1$.

Prima di passare alla simulazione è necessario trovare il tempo di equilibrazione e stabilire la dimensione dei blocchi per il datablock.

### Tempo di equilibrazione

Per determinare il tempo di equilibrazione ho considerato diversi valori dei paramentri $\mu=\{0.9, 1.1, 1.3\}$ e $\sigma=\{0.3,0.5,0.7\}$ e per ciascuna coppia ho osservato l'andamento dell'integranda $f(x)$ con il numero di time step. Di seguito è riportato tale andamento per quei valori di $\mu$ e $\sigma$ che corrispondevano a oscillazioni di $f(x)$ più contenute e attorno a un valore più basso, scelta dei paramentri che più si avvicina a quelli che minimizzano l'energia.

## Ottimizzazione

Ora che tutti i parametri della simulazione sono stati fissati viene eseguita la simulazione Monte Carlo per ogni coppia dei paramentri $\mu$ e $\sigma$ stabiliti prima, usando $10^4$ blocchi. 
Il risultato di ogni simulazione viene stampato nel file finalH.dat. 

Con lo script seguente vengono trovati i valori di $\mu$ e $\sigma$ che minimizzano l'energia della particella.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

mu, sigma, H ,errH = np.loadtxt("finalH.dat", usecols=(0,1,2,3), delimiter='   ', unpack='true')

minH = H[0]
ind = 0
pos = 1
while pos < len(H):
   if H[pos] < minH:
      minH = H[pos]
      ind = pos
   pos = pos + 1

MU = mu[ind]
SIGMA = sigma[ind]
err = errH[ind]

print("Minimum Energy = "+str(minH)+"+/-"+str(err))
print("Mu = "+str(MU)+"   Sigma = "+str(SIGMA))

file = open("opt_parameters.txt", "w")
file.write(str(MU)+"   "+str(SIGMA))
file.close()
```

L'energia assume il valore minimo $E^{\mu _{opt},\sigma _{opt}}_0=-0.4424\pm0.0007$ per $\mu_{opt} = 0.8$ e $\mu_{opt} = 0.6$. Ottenuti questi risultati eseguiamo un'ultreriore simulazione usando $\mu_{opt}$ e $\sigma_{opt}$. Di seguito troviamo l'andamento della stima dell'energia del ground state in funzione del numero dei blocchi. 

```python
blk, progH, errH = np.loadtxt("output.H.0", usecols=(0,2,3), unpack='true') 
plt.errorbar(blk, progH, yerr = errH)
plt.xlabel('#blocks')
plt.ylabel('$<H>$')
plt.grid(True)
plt.title('Variational Montecarlo: MU = '+str(MU)+', SIGMA = '+str(SIGMA))
plt.show()
```

## Confronto con altri metodi numerici

### 1) Differenze finite

Ora vogliamo trovare una soluzione approssimata della funzione d'onda del ground state, e del suo autovalore, integrando numericamente l'equazione di Schrodinger con il metodo delle differenze finite. In seguito vogliamo confrontare la funzione d'onda e l'energia del ground state così ottenuti con quelli stimati precedentemente con il Variational Monte Carlo. 

```python
def Vpot(x):
    return (x**2 - 2.5)*x**2

a = 10
N = 100 # number of iterations
hbar = 1
m = 1

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H) 

# Take the transpose & normalize
psi = np.transpose(psi) 
psi = psi/np.sqrt(dx)       

print("Ground state energy: ", E[0])
#print("1st excited state energy: ", E[1])
#print("2nd excited state energy: ", E[2])
```

Commento su questo risultato energia...

Ora, sempre nel caso di $\mu_{opt}$ e $\sigma_{opt}$, rappresentiamo un grafico che mostri:
- un istogramma delle posizioni campionate $x_i$,
- la funzione $\rho(x) = |\psi ^{\mu_{opt},\sigma_{opt}}(x)|^2/\int dx |\psi ^{\mu_{opt},\sigma_{opt}}(x)|^2$,
- il modulo quadro del ground state ottenuto con le differenze finite $|\psi _{FD}|^2$.

```python
def psif(x):
    return ( np.exp(-((x+MU)**2)/(2*SIGMA**2)) + np.exp(-((x-MU)**2)/(2*SIGMA**2)) ) 

#histogram
sample_x = np.loadtxt("x.dat", usecols=(0), unpack = 'true' )
n_bins = 50
plt.hist(sample_x, n_bins, density = True, range=(-5,5), facecolor='y', label = 'VMC x sample' )
#rho VMC
PSI = psif(x)
norm = 2*SIGMA*np.sqrt(np.pi)*( 1 + np.exp(-(MU**2)/(SIGMA**2)) );
psi2 = (PSI**2)/norm
plt.plot(x,psi2,label = "psi2 VMC",color='red')
#psi^2 FD
plt.plot(x,(psi[0])**2, label = "psi2 FD" ,color='blue')
plt.title("Variational Montecarlo & Finite Differencies")
plt.xlabel('x')
plt.legend()
plt.grid(True)
plt.show()
```

Notiamo che tra la funzione d'onda ottenuta con Variational Montecarlo e quella ottenuta con le differenze finite c'è una leggera discrepanza.

Come funzione di partenza da proiettare sul ground state consideriamo prima la funzione $\psi _1 = 1$ e successivamente $\psi _2 = \psi ^{\mu _{opt},\sigma _{opt}}$. - il profilo dell'istogramma di $x$ ottenuto con PIGS usando $\psi _2$

```python
xq, psi2q, errpsiq = np.loadtxt("probability_psi1.dat", usecols=(0,1,2), unpack = 'true' )
xq_ms, psi2q_ms, errpsiq_ms = np.loadtxt("probability_psiMS.dat", usecols=(0,1,2), unpack = 'true' )

plt.errorbar(xq,psi2q, yerr=errpsiq, label='PIGS PSI_T=1')
plt.errorbar(xq_ms,psi2q_ms, yerr=errpsiq_ms, label='PIGS PSI_T=PSI(MU,SIGMA)')
vmc = psif(xq)
vmc2 = (vmc**2)/norm
plt.plot(xq, vmc2, label='VMC' )
plt.plot(x,(psi[0])**2, label='Finite Differences')
plt.legend()
plt.grid(True)
plt.show()
```

Osservo che le ultime tre curve si sovrappondono mentre la prima il risultato di Variational MonteCarlo si discosta leggermente da essi.

Questa osservazione mi fa sospettare che l'approssimazione della funzione d'onda del ground state fatta con Variational MonteCarlo sia la meno accurata. Probabilmente ciò è dovuto al numero limitato di valori di $\mu$ e $\sigma$ considerati nell'ottimizzazione. Suppongo che considerando un maggior numero di valori in un intorno più ristretto di $\mu _{opt}$ e $\sigma _{opt}$ sia possibile trovare paramentri migliori cui corrisponde una funzione d'onda con un profilo più simile a quello ottenuto con gli altri metodi.
